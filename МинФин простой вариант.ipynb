```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor((
        [3, 12],
        [3, 22],
        [4, 44],
        [4, 32]
    ), dtype=torch.float) #3 на 29 тензор, количество изученных часов и время отведенное на сон

y = torch.tensor((
    [87],
    [67],
    [57],
    [64]), dtype=torch.float) #3 на 1 тензор, оценки

xPredicted = torch.tensor(([3, 11]), dtype=torch.float) #1 на 29 тензор, вход, для которого хотим спрогнозировать оценку
```

```python
class Neu_Net(nn.Module):
    
    def __init__(self, ): #функция инициализации

        super( Neu_Net, self).__init__()
        self.inputSize = 2
        self.outputSize = 1
        self.hiddenSize =30
        
        self.W1 = torch.randn(self.inputSize, self.hiddenSize)
        self.W2 = torch.randn(self.hiddenSize, self.outputSize)
        
    def fw(self, X): #функция форвард
        self.z = torch.matmul(X, self.W1) 
        self.z2 = self.sg(self.z) 
        self.z3 = torch.matmul(self.z2, self.W2)
        o = self.sg(self.z3)
        return o
        
    def sg(self, s):
        return 1 / (1 + torch.exp(-s))
    
    def sgPrime(self, s):
        return s * (1 - s)
    
    def bw(self, X, y, o): #функция Backward
        self.o_error = y - o
        self.o_delta = self.o_error * self.sgPrime(o) # derivative of sig to error
        self.z2_error = torch.matmul(self.o_delta, torch.t(self.W2))
        self.z2_delta = self.z2_error * self.sgPrime(self.z2)
        self.W1 += torch.matmul(torch.t(X), self.z2_delta)
        self.W2 += torch.matmul(torch.t(self.z2), self.o_delta)
        
    def train(self, X, y):
        o = self.fw(X)
        self.bw(X, y, o)
        
    def saveWeights(self, model):
        torch.save(model, "NN")   
             
    def predict(self, net): #вывод результатов прогнозирования
        print ("Прогнозируемые данные на основе трен.весов: ")
        print ("Вход-масштабированный: \n" + str(net))
        print ("Выход: \n" + str(self.fw(net)))
        print( type(self.fw(net)))
        # print ("Вход-масштабированный: \n" + str(xPredicted))
        # print ("Выход: \n" + str(self.fw(xPredicted)))
```

```python
X_max, _ = torch.max(X, 0) #возвращает тензор и индексы
xPredicted_max, _ = torch.max(xPredicted, 0)
X = torch.div(X, X_max)
xPredicted = torch.div(xPredicted, xPredicted_max)
y = y / 100 #так как максимальный результат равен 100
```

```python
NN = Neu_Net()

for i in range(20000):
    # print(f'# {i} ') # Loss: {torch.mean((y - NN(X))**2).detach().item()}')
    NN.train(X, y)

NN.saveWeights(NN)
NN.predict(xPredicted)
```
